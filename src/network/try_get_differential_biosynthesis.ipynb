```python
import pandas as pd
import numpy as np
from glycowork.network.biosynthesis import get_differential_biosynthesis
from glycowork.motif.processing import canonicalize_iupac
```

```python
abundance = pd.read_csv("../../results/data/prepared/processed_abundance.csv", index_col=0)
groups = pd.read_csv("../../results/data/prepared/groups.csv", index_col=0)
structures = pd.read_csv("../../data/glycan_structure_guess_linkage.csv", index_col=0)

data = groups.join(abundance)
data = data[np.isin(data.group, ("HC", "HCC"))]
data.head()
```

```python
prepared = data.drop("group", axis=1)
prepared = prepared.T
prepared.index.name = "glycan"
prepared = structures.join(prepared)
prepared = prepared.rename(columns={"structure": "glycan"})
prepared["glycan"] = [canonicalize_iupac(s) for s in prepared["glycan"]]
prepared = prepared.reset_index(drop=True)
prepared.head()
```

```python
print(len(set(prepared.glycan)))
print(len(prepared.index))
```

```python
group1 = (np.where(data.group == "HC")[0]+1).tolist()
group2 = (np.where(data.group == "HCC")[0]+1).tolist()
```

```python
print(max(group1))
print(max(group2))
print(len(prepared.columns))
```

```python
result = get_differential_biosynthesis(prepared, group1, group2)
```
