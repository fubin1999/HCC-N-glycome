```python
import pandas as pd
```

```python
structures = pd.read_csv("../../data/human_serum_glycans.csv")
```

```python
from glycowork.motif.processing import canonicalize_iupac
glycans = [canonicalize_iupac(ct) for ct in structures['structure']]
```

```python
from glycowork.motif.annotate import annotate_dataset
motifs = annotate_dataset(glycans)
```
